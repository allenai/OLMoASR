```python
import torch
```

```python
import whisper
```

```python
model = whisper.load_model("checkpoints/tiny-en.pt")
```

```python
device = "cuda"
```

```python
checkpoint = torch.load("checkpoints/tiny-en.pt", map_location=device)
```

```python
from open_whisper.model import ModelDimensions
```

```python
dims = ModelDimensions(**checkpoint["dims"])
```

```python
from open_whisper.model import Whisper
```

```python
model = Whisper(dims)
```

```python
model.load_state_dict(checkpoint["model_state_dict"])
```

```python
from collections import OrderedDict
```

```python
import re
```

```python
state_dict = checkpoint["model_state_dict"]
```

```python
model_dict = OrderedDict()
pattern = re.compile("module.")
for k, v in state_dict.items():
    if re.search("module", k):
        model_dict[re.sub(pattern, "", k)] = v
    else:
        model_dict = state_dict
```

```python
model_dict
```

```python
model.load_state_dict(model_dict)
```

```python
audio = whisper.load_audio(
    "data/sanity_check/audio/2CWmrpsN41E/segments/00:00:05.360_00:00:34.800.m4a"
)
audio = whisper.pad_or_trim(audio)

# make log-Mel spectrogram and move to the same device as the model
mel = whisper.log_mel_spectrogram(audio).to(model.device)

options = whisper.DecodingOptions(language="en", without_timestamps=True)

result = whisper.decode(model, mel, options)
```
