```python
from open_whisper import tokenizer, utils
import os
```

```python
text_files = []
for root, dirs, files in os.walk("/home/ubuntu/open_whisper/data/transcripts"):
    if "segments" in root:
        text_files.extend([os.path.join(root, f) for f in os.listdir(root)])
```

```python
text_files
```

```python
def check_text(text_path):
    transcript, *_ = utils.TranscriptReader(file_path=text_path).read()

    if transcript == {}:
        text_tokens = [tokenizer.no_speech]
    else:
        transcript_text = utils.TranscriptReader.extract_text(transcript)

        text_tokens = tokenizer.encode(transcript_text)

    text_tokens = list(tokenizer.sot_sequence_including_notimestamps) + text_tokens

    text_tokens.append(tokenizer.eot)
    if len(text_tokens) > 448:
        return text_path
```

```python
import multiprocessing
```

```python
from tqdm import tqdm
```

```python
tokenizer = tokenizer.get_tokenizer(multilingual=False)
```

```python
with multiprocessing.Pool() as pool:
    out = list(tqdm(pool.imap_unordered(check_text, text_files), total=len(text_files)))
```

```python
out
```

```python
long_texts = [f for f in out if f != None]
len(long_texts)
```
